```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
```

```python
req = requests.get("https://www.ccny.cuny.edu/registrar/fall-2018-academic-calendar")
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
print(soup.prettify)
```

```python
columnhead = ['dates', 'date of the week', 'explanation']
```

```python
all_tables = soup.find_all('table')
```

```python
from IPython.display import HTML
```

```python
HTML(str(all_tables[0]))
```

```python
rows = [row for row in all_tables[0].find_all('tr')]     #table->tr->td
rows[1].find_all('td')
```

```python
value_rows = []
for row in rows[1:]:   #dont want the row with columnhead
    temp_rows = []
    for col in row.find_all('td'):
        temp_rows.append(col.get_text().strip().replace('\n', ' '))    
    value_rows.append(temp_rows)     #temp_rows is a list of row elements of the table

#value_rows is a nested lists now which is a list of every row of table except the row with column head
```

```python
value_rows_tidy = []
for value_row in value_rows:   
    if value_row[-1] != '':
        value_rows_tidy.append(value_row)   #clean the row contains only 'August',September',....
```

```python
value_series =list(zip(*value_rows_tidy))   
```

```python
data_dict = dict(zip(columnhead, value_series))
```

```python
data_dict
```

```python
df = pd.DataFrame(data_dict)
```
